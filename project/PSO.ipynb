```python
class NASModule(nn.Module) :
    def __init__(self, input, output):
        super(NASModule, self).__init__()
        
        self.input = input
        self.output = output
        
        self.fc1 = nn.Conv2d(1,32,3)
        self.fc2 = nn.Conv2d(32,32,3)
        self.fc3 = nn.Conv2d(32,32,3)

    def forward(self,x) :
        out = self.fc3(self.fc2(self.fc1(x)))
        return out
    
```

```python
m = NASModule(1,1)
```

```python
inputs = torch.randn((2,1,28,28))
y = m(inputs)
torchviz.make_dot(y , params = dict(m.named_parameters()))
```
